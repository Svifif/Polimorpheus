```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


k = 10
arr1 = [] 
arr2 = []  

for n in range(1, k + 1):  
    matrix = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            matrix[i][j] = 1 / (j + i + 1)
    
    df = pd.DataFrame(matrix)
    print("Матрица 1:\n", df)
    
    norm1 = np.linalg.norm(matrix, 'fro')
    print("Норма матрицы 1:", norm1)
    

    det = np.linalg.det(matrix)
    
    if det != 0:
        matrix2 = np.linalg.inv(matrix)
        print("Обратная матрица 2:\n", matrix2)
        

        ans = np.linalg.cond(matrix)#хахах оно тут всё время было...
        print("Число обусловленности матрицы:", ans)
        

        arr1.append(n)
        arr2.append(ans)
    else:
        print("Матрица вырождена, число обусловленности не вычислено.")
        arr1.append(n)
        arr2.append(np.nan) 
        
plt.figure(figsize=(10, 6))
plt.grid()
plt.plot(arr1, arr2, marker='o')
plt.title('Зависимость числа обусловленности от размерности матрицы n')
plt.xlabel('Размерность матрицы n')
plt.ylabel('Число обусловленности')
plt.yscale('log')  
plt.xticks(arr1)  
plt.show()
```
