```python
#O(n^3) грустно

import numpy as np
import pandas as pd
import sys
for i in range(3):
    n=5
    A = np.random.randint(0, 15, size=(n, n))
    
    L=np.eye(n,n)
    U=np.zeros((n,n))
    an1 = pd.DataFrame(A)
    print ("Матрица A:\n", an1)
    for i in range(n):
        for j in range(i, n):
            U[i][j] = A[i][j]
            for k in range(i):
                U[i][j] -= L[i][k] * U[k][j]
        L[i][i] = 1  
        for j in range(i + 1, n):
            L[j][i] = A[j][i]
            for k in range(i):
                L[j][i] -= L[j][k] * U[k][i]
            if U[i][i] != 0:
                L[j][i] /= U[i][i]
            else:
               print ("деление на 0")
               sys.exit()
    
    print("Матрица L:\n", pd.DataFrame(L))            
    print("Матрица U:\n", pd.DataFrame(U))
    print("Матрица L*U:")
    print(pd.DataFrame(L@U))
```
